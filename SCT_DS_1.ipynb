```python
import pandas as pd
file_path='/content/API_SP.POP.TOTL_DS2_en_excel_v2_38348.xls'
data = pd.read_excel(file_path, sheet_name='Data', skiprows=3)
data.head()
```

**Filtering relevant columns**


```python
pop_2022 = data[['Country Name', '2022']].dropna()
pop_2022 = pop_2022.sort_values(by='2022' , ascending=False).head(10)
pop_2022
```

**Creating Bar Chart**

```python
import matplotlib.pyplot as plt
plt.bar(pop_2022['Country Name'], pop_2022['2022'] , color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries with Highest Population in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
